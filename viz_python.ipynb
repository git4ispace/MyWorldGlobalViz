```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#creation de donnees
ventes_ferme=[250000,350000,500000,800000,200000]
ventes_poulet=[150000,260000,356000,450000,753000]
annees=['2015','2016','2017','2018','2019']
```

```python
plt.scatter(annees,ventes_ferme, label='Total vente ferme', c='r')
plt.plot(annees,ventes_poulet, label="total vente poulet", c='b')
plt.legend()
plt.xlabel("ANNEES")
plt.ylabel("TOTAL DES VENTES EN FCFA")
plt.title("EVOLUTION DES VENTES DE 2015 A 2019")
plt.show()
```

```python
#vizualisation en deux plan
plt.subplot(2,1,1)
plt.scatter(annees,ventes_poulet, label="total vente poulet", c='lime')

plt.subplot(2,1,2)
plt.plot(annees,ventes_ferme, label="total vente ferme", c='r')
```

```python
#visualisation des donnees en diagramme de bandes
index_annee=np.arange(len(annees))
x=0.25
plt.bar(index_annee,ventes_ferme, label="Total vente ferme", width=0.25)
plt.bar(index_annee+x,ventes_poulet, label="total vente poulet", width=0.25)
plt.legend()
plt.xlabel("ANNEES")
plt.ylabel("TOTAL DES VENTES EN FCFA")
plt.title("EVOLUTION DES VENTES DE 2015 A 2019")
plt.show()
```

```python
#diagramme en secteur pie shart
produit=['P1','P2','P3','P4']
vente=[50000,25000,12000,13000]
plt.pie(vente, labels=produit, autopct='%1.2f%%')
plt.title("EVOLUTION DES VENTES DE 2015 A 2019")
plt.show()
```
